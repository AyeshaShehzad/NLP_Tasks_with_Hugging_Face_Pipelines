#**Text Classification**

###**Sentiment Analysis**

We can classify the text using the Hugging face prebuilt pipelines.

```python
#Import necessary libraries
from transformers import pipeline

#Load sentiment analysis pipeline
sa_pipeline = pipeline("sentiment-analysis")

#perform sentiment analysis
text = "I absolutely loved the movie, it was fantastic!"
result = sa_pipeline(text)
print(result)
```

```python
#Load sentiment anlaysis pipeline
sa_pipeline_1 = pipeline("text-classification")

#Perform sentiment analysis
text_1 = "I absolutely loved the movie, it was fantastic!"
result_1 = sa_pipeline_1(text_1)
print(result_1)
```

Here we noticed that both pipelines are using the same default model "distilbert/distilbert-base-uncased-finetuned-sst-2-english and revision af0f99b", hence both will give the same output.

Instead of relying on the default model, we'll specify a specific one called "roberta-large-mnli" from Hugging Face for our text classification task. This way, we're being explicit about the model we're using rather than relying on the default choice.

```python
from transformers import pipelines
sa_pipeline2 = pipeline(model="roberta-large-mnli")

text2 = 'I eat apple today'
print(sa_pipeline2(text))
```

#**Named Entity Recognition**

**Named Entity Recognition** (NER) is a natural language processing (NLP) task that aims to identify and classify named entities within a text into predefined categories such as person names, organization names, locations, dates, and more. The purpose of NER is to extract and categorize specific entities from unstructured text data, enabling machines to understand the context and relationships between different entities within a document.

###**Why NER?**
Named Entity Recognition (NER) finds applications in various industries where text data analysis is crucial. Here are some industries where NER is commonly used:

**Healthcare and Biomedicine:** In healthcare, NER is used to extract and classify medical entities such as diseases, symptoms, medications, and procedures from electronic health records and biomedical literature.

**Financial Services:** NER is employed in financial services for extracting and categorizing entities such as company names, stock symbols, financial indicators, and regulatory information from news articles, financial reports, and social media feeds.

**Media and Publishing:** NER is used in media and publishing for entity extraction and categorization in news articles, social media posts, and digital content. It enables content recommendation and trend analysis in media monitoring and social listening platforms.

**E-commerce and Retail:** In e-commerce and retail, NER is employed to recognize product names, brands, categories, and attributes from product descriptions, customer reviews, and social media comments. It supports product recommendation, sentiment analysis, and competitive intelligence.

**Government and Public Sector:** NER is utilized by government agencies for extracting and classifying entities such as government organizations, geographic locations, political figures, and policy topics from public documents, legislative texts, and social media discussions. It aids in policy analysis, public opinion monitoring, and disaster response.

**Transportation and Logistics:** In transportation and logistics, NER helps in identifying and categorizing entities such as transportation hubs, geographic locations, shipping companies, and supply chain entities from logistics documents. It supports route optimization, supply chain management, and risk analysis.

```python
#Load transformers
import transformers

#Import pipelines
from transformers import pipeline

#Load named-entity recognition pipeline
ner_pipeline = pipeline('ner')

#Perform named entity recognition
text = "John Smith traveled from New York to Paris, spending $5000 on souvenirs during his trip on April 15th, 2024."
print(ner_pipeline(text))
```

```python
# Import necessary libraries
import transformers  # Library for NLP tasks
from transformers import pipeline  # Module for easy integration of pre-trained models
from diffusers import StableDiffusionPipeline  # Custom pipeline for image generation using Stable Diffusion
import torch  # Library for deep learning computations

# Define the model identifier for the Stable Diffusion model
model_id = "runwayml/stable-diffusion-v1-5"

# Initialize the Stable Diffusion pipeline with the specified model identifier
# Specify torch_dtype=torch.float16 for faster computations on supported hardware
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Move the pipeline to the GPU (if available) for faster image generation
pipe = pipe.to("cuda")

# Define the textual prompt for generating the image
prompt = "a surreal landscape with floating islands and bioluminescent trees"

# Generate the image based on the textual prompt using the Stable Diffusion pipeline
# Access the generated image from the output of the pipeline
image = pipe(prompt).images[0]

# Save the generated image to a file named "astronaut_rides_horse.png"
image.save("astronaut_rides_horse.png")
```
